```python
df['gender']=df['gender'].map({'female':0,'male':1})
```

```python
df['race_ethnicity'].unique()
```

```python
df['race_ethnicity']=df['race_ethnicity'].map({'group A':0,'group B':1,'group C':2,'group D':3,'group E':4})
```

```python
df['parental_level_of_education'].unique()
```

```python
df['parental_level_of_education']=df['parental_level_of_education'].astype('category')
df['encoded_parental_level_of_education']=df['parental_level_of_education'].cat.codes
```

```python
df.head()
```

```python
df.drop('parental_level_of_education',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['lunch']=df['lunch'].map({'standard':0,'free/reduced':1})
```

```python
df['test_preparation_course']=df['test_preparation_course'].map({'none':1,'completed':2})
```

```python
df.head()
```

```python
df.describe()
```

```python
numeric_feature= [feature for feature in df.columns if df[feature].dtype !='o']
categorical_feature= [feature for feature in df.columns if df[feature].dtype =='o']
print("we have {} numerical feature is : {} ".format(len(numeric_feature),numeric_feature))

print("we have {} categorical feature is {}".format(len(categorical_feature),categorical_feature))
```

```python
df['Total_score']=df['math_score']+ df['reading_score']+df['writing_score']
df['Average_score']=df['Total_score']/3
df['Average_score']=df['Average_score'].round(2)
df.head()
```

```python
reading_full=df[df['reading_score']==100]['Average_score'].count()
writing_full=df[df['writing_score']==100]['Average_score'].count()
math_full=df[df['math_score']==100]['Average_score'].count()

print(f"number of student with full marks in reading is {reading_full}")
print(f"number of student with full marks in writing is {writing_full}")
print(f"number of student with full marks in math is {math_full}")
```

```python
fig ,axs= plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='Average_score',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='Average_score',kde=True,hue='gender')
plt.show()
```
